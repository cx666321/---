숙제 1
**굵은 텍스트**


```python
import numpy as np

# 用于存储 3 次彩票结果的列表
lotto_list = []

# 为了保证每次结果不同，但可复现，设置不同的随机种子
np.random.seed(85)

for i in range(3):
    # 生成 1~45 的数组
    numbers = np.arange(1, 46)

    # 随机打乱数组
    np.random.shuffle(numbers)

    # 取前 6 个作为一次彩票结果
    lotto = numbers[:6]

    # 将 NumPy 数组转换为列表并存储
    lotto_list.append(lotto.tolist())

# 将列表转换为 NumPy 二维数组
lotto_array = np.array(lotto_list)

# 输出结果
print("3 次彩票结果构成的二维数组：")
print(lotto_array)

print("数组形状：", lotto_array.shape)
```

```python
# 1. 导入 numpy 库并命名为 np
import numpy as np

# ===============================
# 一、创建随机三维数组（示例参考）
# ===============================

# 指定三维数组的大小
shape_example = (3, 4, 5)

# 创建一个三维数组，元素范围 0~9
arr_example = np.random.randint(0, 10, shape_example)

# 打印三维数组
print("示例三维数组 (3, 4, 5)：")
print(arr_example)


# =========================================
# 二、创建形状为 (3, 3, 6) 的随机三维数组
# =========================================

# 指定三维数组形状
shape_3d = (3, 3, 6)

# 创建随机三维数组，元素范围 0~20
arr_3d = np.random.randint(0, 21, shape_3d)

# 打印结果
print("\n随机三维数组 (3, 3, 6)，元素范围 0~20：")
print(arr_3d)


# =====================================
# 三、创建形状为 (5, 6) 的随机二维数组
# =====================================

# 指定二维数组形状
shape_2d = (5, 6)

# 创建随机二维数组，元素范围 0~100
arr_2d = np.random.randint(0, 101, shape_2d)

# 打印结果
print("\n随机二维数组 (5, 6)，元素范围 0~100：")
print(arr_2d)
```

**숙제 3**

```python

# 1. 导入 numpy 库并命名为 np
import numpy as np

# 2. 使用 arange 创建一维数组 temp
temp = np.arange(48)

print("一维数组 temp：")
print(temp)
print("temp 的形状：", temp.shape)


# =====================================
# 3. 使用 reshape 创建三维数组 (4, 3, 4)
# =====================================

arr_3d = temp.reshape(4, 3, 4)

print("\n三维数组 arr_3d (4, 3, 4)：")
print(arr_3d)
print("arr_3d 的形状：", arr_3d.shape)


# =====================================
# 4. 使用 reshape 创建二维数组 (6, 8)
# =====================================

arr_2d = temp.reshape(6, 8)

print("\n二维数组 arr_2d (6, 8)：")
print(arr_2d)
print("arr_2d 的形状：", arr_2d.shape)
```
